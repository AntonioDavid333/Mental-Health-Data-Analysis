## ANÁLISIS

### SCATTER. CORRELACIÓN entre consumo de frutas y verduras y enfermedades mentales

```python

df_frutas = pd.read_csv("dataframes_limpios/df_frutas.csv")
df_saludMental = pd.read_csv("dataframes_limpios/df_saludMentalEuropa2019.csv")


# df_frutas: Agrupar por país y calcular el consumo promedio
# (Esto resuelve tener múltiples filas por país debido a la columna 'age')
df_frutas_promedio = df_frutas.groupby('country')['value'].mean().reset_index()
df_frutas_promedio.rename(columns={'value': 'Avg_Fruit_Veg_Consumption'}, inplace=True)

# df_saludMental: Calcular el total de enfermedades mentales (igual que antes)
disease_columns = ['Schizophrenia', 'Depression', 'Anxiety', 'Bipolar', 'Eating']
df_saludMental['Total_Mental_Disorders'] = df_saludMental[disease_columns].sum(axis=1)
df_saludMental_preparado = df_saludMental[['country', 'Total_Mental_Disorders']].copy()

# 3. Unir los DataFrames
df_combinado = pd.merge(
    df_frutas_promedio,
    df_saludMental_preparado,
    on='country',
    how='inner'
)

# 4. Calcular la Correlación de Pearson (para entender la relación)
correlation = df_combinado['Avg_Fruit_Veg_Consumption'].corr(
    df_combinado['Total_Mental_Disorders']
)

# 5. Crear el gráfico de dispersión mejorado
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_combinado,
    x='Avg_Fruit_Veg_Consumption',
    y='Total_Mental_Disorders'
)

# Añadir línea de regresión
sns.regplot(
    data=df_combinado,
    x='Avg_Fruit_Veg_Consumption',
    y='Total_Mental_Disorders',
    scatter=False,
    color='red',
    line_kws={'linestyle': '--', 'alpha': 0.7}
)

#anotación de correlación
plt.text(
    x=df_combinado['Avg_Fruit_Veg_Consumption'].max() * 0.95,
    y=df_combinado['Total_Mental_Disorders'].min(),
    s=f'Correlación (ρ): {correlation:.2f}',
    fontsize=12,
    ha='right',
    bbox=dict(facecolor='white', alpha=0.6, edgecolor='gray')
)

#Etiquetas y Título
plt.title('Correlación entre Consumo Promedio de Frutas/Verduras y Total de Trastornos Mentales')
plt.xlabel('Consumo Promedio de Frutas y Verduras (Asumo: % de población con consumo diario)')
plt.ylabel('Suma de Tasas de Prevalencia de Trastornos Mentales')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

print(f"\nCoeficiente de Correlación de Pearson (ρ): {correlation:.2f}")
print("Primeras filas del DataFrame combinado usado para el gráfico:")
print(df_combinado.head(31))
```

```python

trastornos_cols = ['Schizophrenia', 'Depression', 'Anxiety', 'Bipolar', 'Eating']

#Suma de enfermedades
df_saludMentalEuropa2019['Total_Trastornos'] = df_saludMentalEuropa2019[trastornos_cols].sum(axis=1)


df_salud_mental_total = df_saludMentalEuropa2019[['country', 'Total_Trastornos']].copy()
df_salud_mental_total.rename(columns={'country': 'país'}, inplace=True) # Renombrar para facilitar la unión
#cambiamos a pais en vez de country
df_aqi.rename(columns={'country': 'país'}, inplace=True)

#Combinar dataframes
df_combinado_aqi_salud = pd.merge(
    df_aqi[['país', 'aqi_medio']],
    df_salud_mental_total,
    on='país',
    how='inner'
)

plt.figure(figsize=(10, 6))

#Linea de regresion
sns.regplot(
    data=df_combinado_aqi_salud,
    x='aqi_medio',
    y='Total_Trastornos',
    scatter_kws={'alpha': 0.7, 's': 80},
    line_kws={'color': 'red', 'linestyle': '--'} 
)

#Etiquetas de país a los puntitos
for i in range(df_combinado_aqi_salud.shape[0]):
    plt.text(
        df_combinado_aqi_salud['aqi_medio'].iloc[i],
        df_combinado_aqi_salud['Total_Trastornos'].iloc[i],
        '  ' + df_combinado_aqi_salud['país'].iloc[i],
        horizontalalignment='left',
        size='small',
        color='black',
        weight='semibold'
    )

#Etiquetas y Título
plt.title('Correlación entre Calidad del Aire (AQI) y Trastornos Mentales (Europa 2019)')
plt.xlabel('AQI Medio (Índice de Calidad del Aire)')
plt.ylabel('Suma de Tasas de Prevalencia de Trastornos Mentales')
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()

#Calcular y mostrar el Coeficiente de Correlación de Pearson
correlation = df_combinado_aqi_salud['aqi_medio'].corr(df_combinado_aqi_salud['Total_Trastornos'])
print(f"\nCoeficiente de Correlación de Pearson (ρ): {correlation:.2f}")
print("\nPrimeras filas del DataFrame combinado usado para el gráfico:")
print(df_combinado_aqi_salud.head())
```

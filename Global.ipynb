### SCATTER. CORRELACIÓN entre consumo de frutas y verduras y enfermedades mentales

```python

df_frutas = pd.read_csv("dataframes_limpios/df_frutas.csv")
df_saludMental = pd.read_csv("dataframes_limpios/df_saludMentalEuropa2019.csv")


# df_frutas: Agrupar por país y calcular el consumo promedio
# (Esto resuelve tener múltiples filas por país debido a la columna 'age')
df_frutas_promedio = df_frutas.groupby('country')['value'].mean().reset_index()
df_frutas_promedio.rename(columns={'value': 'Avg_Fruit_Veg_Consumption'}, inplace=True)

# df_saludMental: Calcular el total de enfermedades mentales (igual que antes)
disease_columns = ['Schizophrenia', 'Depression', 'Anxiety', 'Bipolar', 'Eating']
df_saludMental['Total_Mental_Disorders'] = df_saludMental[disease_columns].sum(axis=1)
df_saludMental_preparado = df_saludMental[['country', 'Total_Mental_Disorders']].copy()

# 3. Unir los DataFrames
df_combinado = pd.merge(
    df_frutas_promedio,
    df_saludMental_preparado,
    on='country',
    how='inner'
)

# 4. Calcular la Correlación de Pearson (para entender la relación)
correlation = df_combinado['Avg_Fruit_Veg_Consumption'].corr(
    df_combinado['Total_Mental_Disorders']
)

# 5. Crear el gráfico de dispersión mejorado
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_combinado,
    x='Avg_Fruit_Veg_Consumption',
    y='Total_Mental_Disorders'
)

# Añadir línea de regresión
sns.regplot(
    data=df_combinado,
    x='Avg_Fruit_Veg_Consumption',
    y='Total_Mental_Disorders',
    scatter=False,
    color='red',
    line_kws={'linestyle': '--', 'alpha': 0.7}
)

#anotación de correlación
plt.text(
    x=df_combinado['Avg_Fruit_Veg_Consumption'].max() * 0.95,
    y=df_combinado['Total_Mental_Disorders'].min(),
    s=f'Correlación (ρ): {correlation:.2f}',
    fontsize=12,
    ha='right',
    bbox=dict(facecolor='white', alpha=0.6, edgecolor='gray')
)

#Etiquetas y Título
plt.title('Correlación entre Consumo Promedio de Frutas/Verduras y Total de Trastornos Mentales')
plt.xlabel('Consumo Promedio de Frutas y Verduras (Asumo: % de población con consumo diario)')
plt.ylabel('Suma de Tasas de Prevalencia de Trastornos Mentales')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

print(f"\nCoeficiente de Correlación de Pearson (ρ): {correlation:.2f}")
print("Primeras filas del DataFrame combinado usado para el gráfico:")
print(df_combinado.head(31))
```

```python

trastornos_cols = ['Schizophrenia', 'Depression', 'Anxiety', 'Bipolar', 'Eating']

#Suma de enfermedades
df_saludMentalEuropa2019['Total_Trastornos'] = df_saludMentalEuropa2019[trastornos_cols].sum(axis=1)


df_salud_mental_total = df_saludMentalEuropa2019[['country', 'Total_Trastornos']].copy()
df_salud_mental_total.rename(columns={'country': 'país'}, inplace=True) # Renombrar para facilitar la unión
#cambiamos a pais en vez de country
df_aqi.rename(columns={'country': 'país'}, inplace=True)

#Combinar dataframes
df_combinado_aqi_salud = pd.merge(
    df_aqi[['país', 'aqi_medio']],
    df_salud_mental_total,
    on='país',
    how='inner'
)

plt.figure(figsize=(10, 6))

#Linea de regresion
sns.regplot(
    data=df_combinado_aqi_salud,
    x='aqi_medio',
    y='Total_Trastornos',
    scatter_kws={'alpha': 0.7, 's': 80},
    line_kws={'color': 'red', 'linestyle': '--'} 
)

#Etiquetas de país a los puntitos
for i in range(df_combinado_aqi_salud.shape[0]):
    plt.text(
        df_combinado_aqi_salud['aqi_medio'].iloc[i],
        df_combinado_aqi_salud['Total_Trastornos'].iloc[i],
        '  ' + df_combinado_aqi_salud['país'].iloc[i],
        horizontalalignment='left',
        size='small',
        color='black',
        weight='semibold'
    )

#Etiquetas y Título
plt.title('Correlación entre Calidad del Aire (AQI) y Trastornos Mentales (Europa 2019)')
plt.xlabel('AQI Medio (Índice de Calidad del Aire)')
plt.ylabel('Suma de Tasas de Prevalencia de Trastornos Mentales')
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()

#Calcular y mostrar el Coeficiente de Correlación de Pearson
correlation = df_combinado_aqi_salud['aqi_medio'].corr(df_combinado_aqi_salud['Total_Trastornos'])
print(f"\nCoeficiente de Correlación de Pearson (ρ): {correlation:.2f}")
print("\nPrimeras filas del DataFrame combinado usado para el gráfico:")
print(df_combinado_aqi_salud.head())
```

# Relacion Desempleo vs Salud Mental

```python
# Desempleo total por país
df_desempleo_pais = (
    df_desempleo
    .groupby(["ID_Pais", "Pais"], as_index=False)
    .agg(Desempleo_Total=("Desempleo", "sum"))
)

# Variable agregada de salud mental
df_salud["Salud_Mental_Total"] = (
    df_salud[
        ["Schizophrenia", "Depression", "Anxiety", "Bipolar", "Eating"]
    ].sum(axis=1)
)

# Unión de datasets
df_merge = pd.merge(
    df_desempleo_pais,
    df_salud,
    left_on="ID_Pais",
    right_on="Code",
    how="inner"
)
```

```python
# Estadísticas básicas
df_merge[["Desempleo_Total", "Salud_Mental_Total"]].describe()
```

```python
# correlación
corr_total = df_merge["Desempleo_Total"].corr(
    df_merge["Salud_Mental_Total"]
)

print(f"Correlación Desempleo vs Salud Mental Total: {corr_total:.3f}")
```

```python
# Visualización
plt.figure()
plt.scatter(
    df_merge["Desempleo_Total"],
    df_merge["Salud_Mental_Total"]
)
plt.xlabel("Desempleo total por país")
plt.ylabel("Prevalencia total de salud mental")
plt.title(
    f"Desempleo vs Salud Mental en Europa (2019)\n"
    f"Correlación = {corr_total:.2f}"
)
plt.show()
```

```python
# Desempleo juvenil
df_desempleo_joven = df_desempleo[
    df_desempleo["Edad"] == "15 to 24 years old"
]

df_joven_pais = (
    df_desempleo_joven
    .groupby(["ID_Pais", "Pais"], as_index=False)
    .agg(Desempleo_Juvenil=("Desempleo", "sum"))
)

df_merge_joven = pd.merge(
    df_joven_pais,
    df_salud,
    left_on="ID_Pais",
    right_on="Code",
    how="inner"
)

corr_joven = df_merge_joven["Desempleo_Juvenil"].corr(
    df_merge_joven["Salud_Mental_Total"]
)

print(f"Correlación Desempleo Juvenil vs Salud Mental: {corr_joven:.3f}")

plt.figure()
plt.scatter(
    df_merge_joven["Desempleo_Juvenil"],
    df_merge_joven["Salud_Mental_Total"]
)
plt.xlabel("Desempleo juvenil")
plt.ylabel("Prevalencia total de salud mental")
plt.title(
    f"Desempleo Juvenil vs Salud Mental (2019)\n"
    f"Correlación = {corr_joven:.2f}"
)
plt.show()
```

“Aunque intuitivamente se podría esperar una relación directa entre desempleo y problemas de salud mental, el análisis a nivel agregado por país en Europa para 2019 muestra una correlación muy débil. Esto sugiere que otros factores estructurales como los sistemas de bienestar, la sanidad pública o el nivel educativo podrían estar amortiguando el impacto del desempleo en la salud mental.”

```python
# ========================================
# 2. PREPARAR DATOS
# ========================================

# Calcular total de trastornos mentales
df_salud["Total_Mental"] = df_salud[
    ["Schizophrenia", "Depression", "Anxiety", "Bipolar", "Eating"]
].sum(axis=1)

# Normalizar nombres de países para el merge
# (porque uno tiene "Spain" y otro "España")
df_desempleo_sexo_pais["Pais_clean"] = df_desempleo_sexo_pais["Pais"].str.strip()
df_salud["country_clean"] = df_salud["country"].str.strip()
```

```python
# ========================================
# 3. HACER EL MERGE
# ========================================
df_combined = pd.merge(
    df_desempleo_sexo_pais,
    df_salud[["country_clean", "Total_Mental", "Depression", "Anxiety"]],
    left_on="Pais_clean",
    right_on="country_clean",
    how="inner"
)

print("=" * 70)
print("MERGE EXITOSO")
print("=" * 70)
print(f"✓ Países en tabla desempleo: {len(df_desempleo_sexo_pais)}")
print(f"✓ Países en tabla salud mental: {len(df_salud)}")
print(f"✓ Países coincidentes: {len(df_combined)}")
print(f"\nColumnas disponibles:")
print(df_combined.columns.tolist())
```

```python
# ========================================
# 4. ANÁLISIS DE CORRELACIONES
# ========================================
print("\n" + "=" * 70)
print("CORRELACIONES")
print("=" * 70)

correlaciones = {
    "Desempleo Total vs Salud Mental": df_combined["Total"].corr(df_combined["Total_Mental"]),
    "Desempleo Femenino vs Salud Mental": df_combined["Female"].corr(df_combined["Total_Mental"]),
    "Desempleo Masculino vs Salud Mental": df_combined["Male"].corr(df_combined["Total_Mental"]),
    "Brecha de Género vs Salud Mental": df_combined["Brecha (F - M)"].corr(df_combined["Total_Mental"]),
    "Brecha de Género vs Depresión": df_combined["Brecha (F - M)"].corr(df_combined["Depression"]),
    "Brecha de Género vs Ansiedad": df_combined["Brecha (F - M)"].corr(df_combined["Anxiety"])
}

for nombre, valor in correlaciones.items():
    print(f"{nombre:.<50} {valor:>6.3f}")
    
```

```python
# Figura 1: Comparación de correlaciones
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Gráfico 1: Desempleo Total vs Salud Mental
sns.regplot(data=df_combined, x="Total", y="Total_Mental", ax=axes[0, 0],
            scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
axes[0, 0].set_title(f'Desempleo Total vs Salud Mental\nρ = {correlaciones["Desempleo Total vs Salud Mental"]:.3f}')
axes[0, 0].set_xlabel('Desempleo Total')
axes[0, 0].set_ylabel('Prevalencia Salud Mental Total')
axes[0, 0].grid(True, alpha=0.3)

# Gráfico 2: Desempleo por Género
ax2 = axes[0, 1]
x = range(len(df_combined))
width = 0.35
ax2.scatter(df_combined["Female"], df_combined["Total_Mental"], 
           alpha=0.6, label='Femenino', s=100, color='pink')
ax2.scatter(df_combined["Male"], df_combined["Total_Mental"], 
           alpha=0.6, label='Masculino', s=100, color='lightblue')
ax2.set_xlabel('Desempleo')
ax2.set_ylabel('Salud Mental Total')
ax2.set_title('Desempleo por Género vs Salud Mental')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Gráfico 3: Brecha de Género vs Salud Mental
sns.regplot(data=df_combined, x="Brecha (F - M)", y="Total_Mental", ax=axes[1, 0],
            scatter_kws={'alpha': 0.6, 'color': 'purple'}, line_kws={'color': 'darkred'})
axes[1, 0].set_title(f'Brecha de Género (F-M) vs Salud Mental\nρ = {correlaciones["Brecha de Género vs Salud Mental"]:.3f}')
axes[1, 0].set_xlabel('Brecha de Género en Desempleo')
axes[1, 0].set_ylabel('Prevalencia Salud Mental Total')
axes[1, 0].grid(True, alpha=0.3)

# Gráfico 4: Top 10 países con mayor brecha
top10_brecha = df_combined.nlargest(10, "Brecha (F - M)").sort_values("Brecha (F - M)")
axes[1, 1].barh(top10_brecha["Pais"], top10_brecha["Brecha (F - M)"], color='coral')
axes[1, 1].set_xlabel('Brecha de Género (F - M)')
axes[1, 1].set_title('Top 10: Mayor Brecha de Género en Desempleo')
axes[1, 1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ========================================
# 6. ANÁLISIS ESPECÍFICO: DEPRESIÓN Y ANSIEDAD
# ========================================

fig2, axes2 = plt.subplots(1, 2, figsize=(15, 6))

# Depresión vs Brecha de Género
sns.regplot(data=df_combined, x="Brecha (F - M)", y="Depression", ax=axes2[0],
            scatter_kws={'alpha': 0.6, 'color': 'steelblue'})
axes2[0].set_title(f'Brecha de Género vs Depresión\nρ = {correlaciones["Brecha de Género vs Depresión"]:.3f}')
axes2[0].set_xlabel('Brecha de Género (F - M)')
axes2[0].set_ylabel('Prevalencia de Depresión')
axes2[0].grid(True, alpha=0.3)

# Ansiedad vs Brecha de Género
sns.regplot(data=df_combined, x="Brecha (F - M)", y="Anxiety", ax=axes2[1],
            scatter_kws={'alpha': 0.6, 'color': 'orange'})
axes2[1].set_title(f'Brecha de Género vs Ansiedad\nρ = {correlaciones["Brecha de Género vs Ansiedad"]:.3f}')
axes2[1].set_xlabel('Brecha de Género (F - M)')
axes2[1].set_ylabel('Prevalencia de Ansiedad')
axes2[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```

## ANÁLISIS

### SCATTER. CORRELACIÓN entre consumo de frutas y verduras y enfermedades mentales

```python

trastornos_cols = ['Schizophrenia', 'Depression', 'Anxiety', 'Bipolar', 'Eating']

#Suma de enfermedades
df_salud['Total_Trastornos'] = df_salud[trastornos_cols].sum(axis=1)


df_salud_mental_total = df_salud[['country', 'Total_Trastornos']].copy()
df_salud_mental_total.rename(columns={'country': 'país'}, inplace=True) # Renombrar para facilitar la unión
#cambiamos a pais en vez de country
df_aqi.rename(columns={'country': 'país'}, inplace=True)

#Combinar dataframes
df_combinado_aqi_salud = pd.merge(
    df_aqi[['país', 'aqi_medio']],
    df_salud_mental_total,
    on='país',
    how='inner'
)

plt.figure(figsize=(10, 6))

#Linea de regresion
sns.regplot(
    data=df_combinado_aqi_salud,
    x='aqi_medio',
    y='Total_Trastornos',
    scatter_kws={'alpha': 0.7, 's': 80},
    line_kws={'color': 'red', 'linestyle': '--'} 
)

#Etiquetas de país a los puntitos
for i in range(df_combinado_aqi_salud.shape[0]):
    plt.text(
        df_combinado_aqi_salud['aqi_medio'].iloc[i],
        df_combinado_aqi_salud['Total_Trastornos'].iloc[i],
        '  ' + df_combinado_aqi_salud['país'].iloc[i],
        horizontalalignment='left',
        size='small',
        color='black',
        weight='semibold'
    )

#Etiquetas y Título
plt.title('Correlación entre Calidad del Aire (AQI) y Trastornos Mentales (Europa 2019)')
plt.xlabel('AQI Medio (Índice de Calidad del Aire)')
plt.ylabel('Suma de Tasas de Prevalencia de Trastornos Mentales')
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()

#Calcular y mostrar el Coeficiente de Correlación de Pearson
correlation = df_combinado_aqi_salud['aqi_medio'].corr(df_combinado_aqi_salud['Total_Trastornos'])
print(f"\nCoeficiente de Correlación de Pearson (ρ): {correlation:.2f}")
print("\nPrimeras filas del DataFrame combinado usado para el gráfico:")
print(df_combinado_aqi_salud.head())
```

# Relacion Desempleo vs Salud Mental

```python
# LIMPIEZA DF SALUD MENTAL
df_salud = df_salud.rename(columns={
    "country": "Pais",
    "Year": "Año"
})

df_salud = df_salud.drop(columns=["Unnamed: 0"], errors="ignore")
```

```python
# Desempleo total por país
desempleo_pais_anio = (
    df_desempleo
    .groupby(["Pais", "Año"], as_index=False)["Desempleo"]
    .sum()
)

df_analisis = pd.merge(
    df_salud,
    desempleo_pais_anio,
    on=["Pais", "Año"],
    how="inner"
)

df_analisis.head()
```

```python
# Preparación del DataFrame para el análisis de correlación
df_modelo = df_analisis[[
    "Pais",
    "Año",
    "Desempleo",
    "Depression",
    "Anxiety"
]]
df_modelo = df_modelo.dropna(subset=["Desempleo", "Depression"])
df_modelo.head()
```

```python
# Estadísticas básicas
df_modelo.describe()
```

```python
# correlación
corr_desemp_dep = df_modelo["Desempleo"].corr(df_modelo["Depression"])

print(f"Correlación Desempleo vs Depresión: {corr_desemp_dep:.3f}")
```

```python
corr_desemp_anx = df_modelo["Desempleo"].corr(df_modelo["Anxiety"])

print(f"Correlación Desempleo vs Ansiedad: {corr_desemp_anx:.3f}")
```

```python
# Visualización
#Desempleo vs Depresión
sns.regplot(
    data=df_modelo,
    x="Desempleo",
    y="Depression",
    scatter_kws={"alpha": 0.6}
)

plt.title("Relación entre desempleo y depresión")
plt.xlabel("Desempleo total")
plt.ylabel("Depresión (%)")
plt.show()
```

```python
#Desempleo vs Ansiedad
sns.regplot(
    data=df_modelo,
    x="Desempleo",
    y="Anxiety",
    scatter_kws={"alpha": 0.6}
)

plt.title("Relación entre desempleo y ansiedad")
plt.xlabel("Desempleo total")
plt.ylabel("Ansiedad (%)")
plt.show()
```

```python
df_modelo["Nivel_Desempleo"] = pd.qcut(
    df_modelo["Desempleo"],
    q=3,
    labels=["Bajo", "Medio", "Alto"],
    duplicates="drop"
)
sns.boxplot(
    data=df_modelo,
    x="Nivel_Desempleo",
    y="Depression"
)

plt.title("Depresión según nivel de desempleo")
plt.xlabel("Nivel de desempleo")
plt.ylabel("Depresión (%)")
plt.show()
```

```python
# 1. Convertir la lista a DataFrame
df_poblacion = pd.DataFrame(paises_europa_poblacion)
df_desempleo.reset_index(drop=True, inplace=True)
df_desempleo['Desempleo_Pct'] = (
    df_desempleo['Desempleo'] * 100 / 
    df_desempleo['Pais'].map(df_poblacion.set_index('Pais')['poblacion'])
)
df_desempleo['Desempleo_Pct'] = df_desempleo['Desempleo_Pct'] * 100

df_desempleo.head()
```

```python
#Eliminar la primera columna de salud
df_salud = df_salud.drop(columns=df_salud.columns[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# CALCULAR EL PROMEDIO DE ENFERMEDADES MENTALES
enfermedades = ['Schizophrenia', 'Depression', 'Anxiety', 'Bipolar', 'Eating']
df_salud['Prevalence'] = df_salud[enfermedades].mean(axis=1)

# PREPARAR LOS DATAFRAMES
df_salud = df_salud.rename(columns={'Year': 'year'})
df_frutas = df_frutas.rename(columns={'time': 'year'})
df_desempleo = df_desempleo.rename(columns={'Año': 'year'})

# AGRUPAR POR AÑO
evolucion_salud = df_salud.groupby('year')['Prevalence'].mean().reset_index()
evolucion_frutas = df_frutas.groupby('year')['value'].mean().reset_index()
evolucion_desempleo = df_desempleo.groupby('year')['Desempleo_Pct'].mean().reset_index()

# UNIR DATAFRAMES
df_plot = evolucion_salud.merge(evolucion_frutas, on='year', how='outer')
df_plot = df_plot.merge(evolucion_desempleo, on='year', how='outer')

# FILTRO 2014–2019
df_plot = df_plot[(df_plot['year'] >= 2014) & (df_plot['year'] <= 2019)]

# CONFIGURACIÓN DEL GRÁFICO
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

sns.lineplot(data=df_plot, x='year', y='Prevalence', marker='o',
             label='Salud Mental (Promedio Enfermedades %)', linewidth=2.5)
sns.lineplot(data=df_plot, x='year', y='value', marker='s',
             label='Consumo Frutas (%)', linewidth=2.5)
sns.lineplot(data=df_plot, x='year', y='Desempleo_Pct', marker='^',
             label='Desempleo (%)', linewidth=2.5)

plt.title('Evolución de Indicadores Sociales (2014 - 2019)', fontsize=16, pad=20)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Valor Promedio', fontsize=12)
plt.xticks(range(2014, 2020))
plt.legend(title='Indicadores', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
